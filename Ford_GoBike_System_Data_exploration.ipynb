## Univariate Exploration

I'll start by looking at the distribution of the main variable of interest: duration_min.

```python
# Distribution of the duration_min feature

fig, ax = plt.subplots(nrows= 1, ncols= 2, figsize= (15,5))
bin_size = 3
bins = np.arange(0, data.duration_min.max() + bin_size, bin_size)
ax[0].hist(data.duration_min, bins= bins)
ax[0].set_xlim([0, 300])
ax[0].set_xlabel('Trip duration (Minutes)')

sns.boxplot(data.duration_min, ax= ax[1]);
```

```python
data.duration_min.describe()
```

The distribution is highly right_skewed with several extreme values. So, we will consider trip durations greater than the maximum whisker (Q3 + 1.5 IQR) as outliers and we will drop them out.

```python
# Calculate Q3
duration_q3 = data.duration_min.quantile(0.75)
# Calculate IQR
duration_IQR = data.duration_min.quantile(0.75) - data.duration_min.quantile(0.25)
# Get the index of outliers
outliers_indx = data.query('duration_min > @duration_q3 + @duration_IQR * 1.5').index
# Drop the outliers
data.drop(index= outliers_indx, inplace= True)
```

```python
# Distribution of the duration_sec feature after removing outliers

fig, ax = plt.subplots(nrows= 1, ncols= 2, figsize= (15,5))
bin_size = 0.5
bins = np.arange(0, data.duration_min.max() + bin_size, bin_size)
ax[0].hist(data.duration_min, bins= bins)
ax[0].set_xlabel('Trip duration (Minutes)')

sns.boxplot(data.duration_min, ax= ax[1]);
```

```python
# Distribution of the duration_sec feature after log transformation
bin_size = 0.05
bins = 10 ** np.arange(0, np.log10(data.duration_min.max()) + bin_size, bin_size)
plt.figure(figsize= (8,5))
plt.hist(data.duration_min, bins= bins)
plt.xscale('log')
plt.xticks([0.1, 0.3, 1, 3, 10, 30, 100, 300, 1000], ['0.1', '0.3', '1', '3', '10', '30', '100', '300', '1000'])
plt.xlabel('Trip duration (Minutes)');
```

The trip duration distribution is heavily right-skewed, with many trips on the short duration end, and few trips on the long duration trips. When plotted on a log-scale the trip duration distribution looks unimodal, with one large peak around 10 minutes.

```python
catg_vars = ['day_of_week', 'user_type', 'member_gender']

# Creat a box plot for each categorical variable vs. duration_min
def boxgrid(x, y, **kwargs):
    default_color = sns.color_palette()[0]
    sns.boxplot(x, y, color = default_color)

g = sns.PairGrid(data = data, y_vars = ['duration_min'], x_vars = catg_vars, size= 6)
g.map(boxgrid)
plt.show();
```

The above box plots shows that:<br/>
